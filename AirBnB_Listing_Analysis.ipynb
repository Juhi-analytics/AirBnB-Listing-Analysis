```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
listings = pd.read_csv("Listings.csv",encoding="ISO-8859-1", engine='python', on_bad_lines='skip')
```

```python
listings.head()
```

```python
listings.info()
```

```python
listings['host_since']=pd.to_datetime(listings['host_since'])
listings.info()
```

```python
Paris_listings = (listings.query("city =='Paris'").loc[:,['host_since','city','accommodates','price','neighbourhood']])
Paris_listings.info()
```

```python
Paris_listings.isna().sum()
```

```python
Paris_listings.describe()
```

```python
Paris_listings.query("accommodates==0 and price==0").count()
```

Prepare for visualization

```python
Paris_listings_neighbourhood = (
    Paris_listings
    .groupby('neighbourhood')
    .agg({'price':"mean"})
    .sort_values('price')
)


Paris_listings_neighbourhood.tail()
```

```python
Paris_listings_accommodates = (
    Paris_listings
    .query("neighbourhood == 'Elysee'")
    .groupby('accommodates')
    .agg({"price":"mean"})
    .sort_values('price')
)
Paris_listings_accommodates.head()
```

```python
Paris_listings_over_time = (
    Paris_listings
    .set_index("host_since")
    .resample("Y")
    .agg({
        "neighbourhood":"count" ,
        "price":"mean"
    })
)
Paris_listings_over_time.head()
```

```python
(Paris_listings_neighbourhood
 .plot
 .barh(
     title="Average Listing Price by Paris Neighbourhood",
     xlabel="Price per night(Euros)",
     ylabel="Neighbourhood",
     color='red',
     legend=None
 )
)
sns.despine()
```

```python
(Paris_listings_accommodates
 .plot
 .barh(
     title="Average Listing Price by Acommodation Number ",
     xlabel="Price per night(Euros)",
     ylabel="Acommodation Capacity",
     color='red',
     legend=None
 )
)
sns.despine()
```

```python
Paris_listings_over_time['neighbourhood'].plot(
    ylabel='New Hosts',
    title='New Airbnb Host Paris over Time',
    color='red'
)
sns.despine()
```

```python
Paris_listings_over_time['price'].plot(
    ylabel="Average Price(Euros)",
    title='Average Airbnb Price in Paris over Time',
    color='red'
)
sns.despine()
```

```python
fig, ax =plt.subplots()
ax.plot(
    Paris_listings_over_time.index,
    Paris_listings_over_time['neighbourhood'],
    label="New Hosts",
    c='red'
)
ax.set_ylabel('New Host')

ax2=ax.twinx()
ax2.plot(
    Paris_listings_over_time.index,
    Paris_listings_over_time['price'],
    label="Average Price",
)
ax2.set_ylim(0)
ax2.set_ylabel('Average Price ')
ax.set_title("2015 Regulations Lead to Fewer New Hosts, Higher Prices")
```
